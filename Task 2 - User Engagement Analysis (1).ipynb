```python
data = telcom.copy()
data['Total Traffic (Bytes)']= data['Total DL (Bytes)'] + data["Total UL (Bytes)"]
data["Total Youtube Volume(Bytes)"]= data["Youtube DL (Bytes)"] + data["Youtube UL (Bytes)"]
data["Total Google Volume(Bytes)"]= data["Google DL (Bytes)"] + data["Google UL (Bytes)"]
data["Total Email Volume(Bytes)"]= data["Email DL (Bytes)"] + data["Email UL (Bytes)"]
data["Total Social Media Volume(Bytes)"]= data["Social Media DL (Bytes)"] + data["Social Media UL (Bytes)"]
data["Total Netflix Volume(Bytes)"]= data["Netflix DL (Bytes)"] + data["Netflix UL (Bytes)"]
data["Total Gaming Volume(Bytes)"]= data["Gaming DL (Bytes)"] + data["Gaming UL (Bytes)"]
data["Total Other Volume(Bytes)"]= data["Other DL (Bytes)"] + data["Other UL (Bytes)"]
```

```python
## Aggregate total data volume (in Bytes) during this session for each application
apps_data_volume = data.groupby(['Dur. (ms)','IMSI', 'MSISDN/Number']).agg({
    'Total Traffic (Bytes)': 'sum',
    'Total Youtube Volume(Bytes)': 'sum',
    'Total Google Volume(Bytes)': 'sum',
    'Total Email Volume(Bytes)': 'sum',
    'Total Social Media Volume(Bytes)': 'sum',
    'Total Netflix Volume(Bytes)': 'sum',
    'Total Gaming Volume(Bytes)': 'sum',
    'Total Other Volume(Bytes)': 'sum',
}).reset_index()

apps_data_volume.head()
```

### Sessions frequency

```python
# Calculate the sessions frequency for each user
sessions_frequency = telcom['MSISDN/Number'].value_counts()

sessions_frequency = pd.DataFrame({'Sessions Frequency': sessions_frequency}).reset_index()

# Display the head of the sessions frequency series
sessions_frequency.head()
```

## The Duration of the session

```python
# Calculate the duration of each session
session_durations = telcom.groupby('MSISDN/Number')['Dur. (ms)'].sum()

## Convert the result into a DataFrame
session_durations = pd.DataFrame({'Sessions Duration (ms)': session_durations}).reset_index()

# Display the head of the session durations series
session_durations.head(10)
```

***The table above shows the total duration of sessions for the top five users in the dataset, measured in milliseconds. Each row represents a unique user identified by their MSISDN/Number and the corresponding sum of durations for all sessions they had.***

## The session total traffic (download and upload (bytes))


```python
# Calculate the total traffic (download and upload in bytes) for each session
session_total_traffic = telcom.groupby('MSISDN/Number')[['Total DL (Bytes)', 'Total UL (Bytes)']].sum()

# Add a new column for the total traffic per session
session_total_traffic['Total Traffic (Bytes)'] = session_total_traffic['Total DL (Bytes)'] + session_total_traffic['Total UL (Bytes)']

# Display the head of the session total traffic dataframe
session_total_traffic
```

***The table displays the total download and upload traffic for the top five users, along with their combined total traffic.***
